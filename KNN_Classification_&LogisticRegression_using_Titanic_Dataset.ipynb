Dataset link: https://www.kaggle.com/competitions/titanic

```python
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
```

```python
data.shape # size of data
```

```python
data.head()
```

```python
data.info
```

```python
data.isnull().sum()
```

```python
data=data.drop(["PassengerId","Pclass","Name","Cabin","Ticket"],axis=1)
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
cols=['SibSp','Parch','Fare']
for col in cols:
  data[col].fillna(data[col].median,inplace=True)
```

```python
data['Embarked'].fillna("U",inplace=True)
```

```python
data['Sex']=data['Sex'].map({'male':1,'female':0})


data['Embarked']=data['Embarked'].map({'S':0,'C':1,'Q':2,'U':3})
```

```python

X=data[['Sex','Age','SibSp','Parch','Fare','Embarked']]
y=data['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=0)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model= KNeighborsClassifier(n_neighbors=4)
knn_model.fit(X_train,y_train)

y_pred = knn_model.predict(X_test)
```

```python
pd.DataFrame({'Y_Test':y_test, 'Y_Pred': y_pred})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression()
logreg_model.fit(X_train, y_train)

y_pred = logreg_model.predict(X_test)
```

```python
pd.DataFrame({'Y_Test':y_test, 'Y_Pred': y_pred})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
coefficients = logreg_model.coef_
print(coefficients)
```
