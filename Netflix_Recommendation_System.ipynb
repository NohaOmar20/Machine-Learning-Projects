# Netflix Movie Rating Dataset

Dataset link: https://www.kaggle.com/datasets/rishitjavia/netflix-movie-rating-dataset

```python
import pandas as pd
```

```python
movies=pd.read_csv('/content/Netflix_Dataset_Movie.csv')
movies.head()
```

```python
ratings=pd.read_csv('/content/Netflix_Dataset_Rating.csv')
ratings.head()
```

```python
movie_ratings=pd.merge(movies,ratings)

movie_ratings.head()
```

```python
movie_ratings.tail()
```

```python
movie_ratings=movie_ratings.pivot_table(index=['User_ID'],columns=['Name'],values='Rating')
movie_ratings.head()
```

```python
movies['Name'][movies['Name'].str.contains('Wild')]
```

```python
movieName=movie_ratings['Character']
movieName
```

```python
similar_movies=movie_ratings.corrwith(movieName)

similar_movies.head(10)
```

```python
similar_movies=movie_ratings.corrwith(movieName)
similar_movies=similar_movies.dropna()

df=pd.DataFrame(similar_movies)

df.head(10)
```
